# Exercise 7

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
# Group transactions by Member_number and Date
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).tolist()
```

### 1. Apply one-hot encoding

```python
# Apply one-hot encoding using TransactionEncoder
one_hot_transformer = TransactionEncoder()
encoded_array = one_hot_transformer.fit_transform(client_basket)
```

```python
# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=one_hot_transformer.columns_)
```

```python
# Display first few rows of encoded data
print("One-hot encoded transaction data:")
print(encoded_df.head())
```

```python
encoded_df.head()
```

```python
encoded_df.shape
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
frequent_itemsets = apriori(encoded_df, min_support=0.0005, use_colnames=True)
```

```python
print("\nFrequent itemsets found using Apriori:")
print(frequent_itemsets.head())
```

3. Display the association rules using the lift metric

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
print("\nAssociation Rules (Sorted by Lift):")
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by="lift", ascending=False).head()
```
